# Twitter Scraper

This script is for scraping tweets from Twitter with the **tweepy** library.

- **pip**,
- **pandas**

### Python Script

The next couple of cells are an example of calling the Twitter API, collecting tweets into a list, turning the list into a Pandas dataframe, and outputing the dataframe as a CSV.

```python
###
'''
Title: Twitter Scraper script
Purpose: To scrape tweets rom Twitter

'''

### Dependencies ###
'''
# Install Tweepy library

python -m pip install tweepy

'''
## Import tweepy and re libraries
import tweepy
import re

from tweepy import OAuthHandler

### Declare Consumer API and Access Token
consumer_api_key = '...'
consumer_api_secret = '...' 
access_token = '...'
access_token_secret ='...'

### call OAuthHandler to connect to Twitter API

#### OAuthhandler takes our Consumer API
authorizer = OAuthHandler(consumer_api_key, consumer_api_secret)
authorizer.set_access_token(access_token, access_token_secret)

### fetch and collect tweets into a list
### returns an object called SearchResults

api = tweepy.API(authorizer, timeout=15)
 
all_tweets = []
 
search_query = '...'
 
for tweet in tweepy.Cursor(api.search, q=search_query, lang='en', result_type='recent').items(200):
    
    all_tweets.append(tweet.text)





```

```python
### Saving information to a Pandas Dataframe, making a visualization (may need to adjust), and exporting as a CSV

import pandas as pd
import plotly.express as px


df = pd.DataFrame(all_tweets)
'''
fig = px.histogram(df, x="all_tweets")
fig.show()
'''

df.to_csv("tweets.csv")
```
